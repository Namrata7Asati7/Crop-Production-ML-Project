```python
X = data_set.drop(columns=['State_Name','Crop_Type','Area_in_hectares','Production_in_tons','Crop'], axis = 1)
Y = data_set['Crop']
```

```python
# Checking whether the dataset is balanced or not

K_Counts = X["K"].value_counts()
temp = pd.DataFrame({
    "Potassium": K_Counts.index,
    "Counts":K_Counts.values
})

plt.figure(figsize = (18,8))
sns.barplot(x = "Potassium", y = "Counts", data = temp)
plt.xticks(rotation=90)
plt.show()
```

```python
X_train, X_test,Y_train,Y_test = train_test_split(X, Y, test_size=.2, stratify=Y, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

rf = RandomForestClassifier(n_estimators=100)
```

```python
# Fit the model on the training data
rf.fit(X_train, Y_train)

# Predict the target for the test set
y_pred1 = rf.predict(X_test)
# y_pred2 = rf.predict(Y_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
import matplotlib.pyplot as plt
cf_matrix = confusion_matrix(Y_test, y_pred1)
plt.figure(figsize=(12,8))
sns.heatmap(cf_matrix, annot=True)
plt.title("Confusion Matrix for Random Forest Classifier on Test Data")
plt.show()
```

```python
accuracy = accuracy_score(Y_test, y_pred1)

print(f'Accuracy: {accuracy}')
```
